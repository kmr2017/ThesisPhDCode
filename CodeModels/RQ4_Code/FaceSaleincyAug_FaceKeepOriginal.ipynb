 #  **ISS_cross**

```python
import pandas as pd
import numpy as np
import os
paths = ['/content/drive/MyDrive/Work_to_do/Face Dataset/Results/profession_wise_iss/']
overall_intra=[]
overall_cross = []
for pth in paths:
  Intra= []
  Cross= []
  count=0
  # print(os.listdir(pth))
  ls =  []
  print(pth)
  for fl in os.listdir(pth):
    count+=1
    df = pd.read_csv(pth+fl)
    ls.append(fl)
    Intra.append(df['ImgSim_Intra'].values)
    Cross.append(df['ImgSim_Cross'].values)

    # print(df['Query'].values, df['ImgSim_Intra'].values, df['ImgSim_Cross'].values)
    if count%5==0:
      # print(ls)
      ls =  []
      Intra = np.array(Intra)
      Cross = np.array(Cross)
      # print(Intra)
      print(df['Query'].values[0])
      # print(np.round(np.mean(Intra,axis=0), 6))
      print(np.round(np.mean(Intra), 4))
      print(np.round(np.mean(Cross),4))
      overall_intra.append(np.round(np.mean(Intra), 4))
      overall_cross.append(np.round(np.mean(Cross),4))
      Intra= []
      Cross= []
print('Overall Mean')
print(round(np.mean(overall_intra),4), round(np.mean(overall_cross),4))
```
